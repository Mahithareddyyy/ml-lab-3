```python
import pandas as pd
import numpy as np
df=pd.read_excel('embeddingdata.xlsx')
df
```

```python
class_0_df = df[df['Label'] == 0]
class_1_df = df[df['Label'] == 1]
class_2_df = df[df['Label'] == 2]
class_3_df = df[df['Label'] == 3]
class_4_df = df[df['Label'] == 4]
class_5_df = df[df['Label'] == 5]
print("class0 data:")
class_0_df
```

```python
class_0_arr = class_0_df.drop(columns=['Label']).values
class_1_arr = class_1_df.drop(columns=['Label']).values
```

```python
#finding centroid of classes with label value 0 and 1
centroid_class_0 = np.mean(class_0_arr, axis=0)
centroid_class_1 = np.mean(class_1_arr, axis=0)
print("Centroid of class0 is:")
print(centroid_class_0)
print("Centroid of class1 is:")
print(centroid_class_1)
```

```python
#finding standard deviation of classes with label value 0 and 1
sd_class_0 = np.std(class_0_arr, axis=0)
sd_class_1 = np.std(class_1_arr, axis=0)
print("Standard deviation of class0 is:")
print(sd_class_0)
print("Standard deviation of class1 is:")
print(sd_class_1)
```

```python
from scipy.spatial.distance import euclidean
```

```python
#calculating interclass distance
euclidean_distance = euclidean(centroid_class_0, centroid_class_1)
print("Distance between centroids i.e interclass distance is:")
print(euclidean_distance)
```

```python
#calculating intradistance of class_0
intradistance_class_0 = np.mean(np.linalg.norm(class_0_arr - centroid_class_0, axis=1))
```

```python
import pandas as pd
vector1 = df.loc[0, ["embed_1", "embed_2"]].values
vector2 = df.loc[1, ["embed_1", "embed_2"]].values
print("Feature Vector 1:")
print(vector1)
print("\nFeature Vector 2:")
print(vector2)
```

```python
def minkowski_distance(x, y, r):
    return np.sum(np.abs(x - y) ** r) ** (1 / r)
values_r = range(1, 11)
distances = [minkowski_distance(vector1, vector2, r) for r in values_r]
print(distances)
plt.figure(figsize=(10, 7))
plt.plot(values_r, distances, marker='o', linestyle='-')
plt.title("Minkowski Distance vs Values of r")
plt.xlabel("r")
plt.ylabel("Minkowski Distance")
plt.grid(True)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
X = np.vstack((class_0_arr, class_1_arr))
y = np.hstack((np.zeros(class_0_arr.shape[0]), np.ones(class_1_arr.shape[0])))
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=40)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(X, y)
```

```python
accuracy = neigh.score(X_test, y_test)
print(f"Accuracy on the test set: {accuracy:2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
accuracies_kNN = []
accuracies_NN = []
#k varies from 1 to 11
k_values = range(1, 12)
for k in k_values:
    # Train the kNN classifier with the current k value
    kNN_classifier = KNeighborsClassifier(n_neighbors=k)
    kNN_classifier.fit(X_train, y_train)
    y_pred_kNN = kNN_classifier.predict(X_test)
    # Calculating accuracy for kNN classifier
    accuracy_kNN = accuracy_score(y_test, y_pred_kNN)
    accuracies_kNN.append(accuracy_kNN)
    if k == 1:
        # Train the NN classifier for k=1
        NN_classifier = KNeighborsClassifier(n_neighbors=1)
        NN_classifier.fit(X_train, y_train)
        y_pred_NN = NN_classifier.predict(X_test)
        # Calculating accuracy for NN where k=1
        accuracy_NN = accuracy_score(y_test, y_pred_NN)
        accuracies_NN.append(accuracy_NN)
# Creating an accuracy plot
plt.figure(figsize=(15, 8))
plt.plot(k_values, accuracies_kNN, marker='o', label='kNN (k=3)')
plt.plot([1], accuracies_NN, marker='o', label='NN (k=1)')
plt.title('Accuracy vs. k Value')
plt.xlabel('k Value')
plt.ylabel('Accuracy')
plt.xticks(k_values)
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
y_test_pred = neigh.predict(X_test)
conf_matrix_test = confusion_matrix(y_test, y_test_pred)
print(conf_matrix_test)
tp = np.sum(np.diag(conf_matrix_test))
tn = np.sum(conf_matrix_test) - (np.sum(np.sum(conf_matrix_test, axis=0)) - tp)  
fp = np.sum(conf_matrix_test, axis=0) - np.diag(conf_matrix_test)
fn = np.sum(conf_matrix_test, axis=1) - np.diag(conf_matrix_test)
print("Value of TP is :", tp)
print("Value of TN is :", tn)
print("Value of FP is :", np.sum(fp))
print("Value of FN is :", np.sum(fn))


```

```python
p=np.sum(fn)
q=tp+p
recall=tp/q
print("Value of recall is: ",recall)
```
