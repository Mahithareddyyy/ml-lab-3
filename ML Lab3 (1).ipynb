```python
import pandas as pd
import numpy as np
df=pd.read_excel('embeddingdata.xlsx')
df
```

```python
class_0_df = df[df['Label'] == 0]
class_1_df = df[df['Label'] == 1]
class_2_df = df[df['Label'] == 2]
class_3_df = df[df['Label'] == 3]
class_4_df = df[df['Label'] == 4]
class_5_df = df[df['Label'] == 5]
print("class0 data:")
class_0_df
```

```python
print("class1 data:")
class_1_df
```

```python
class_0_arr = class_0_df.drop(columns=['Label']).values
class_1_arr = class_1_df.drop(columns=['Label']).values
```

```python
#finding centroid of classes with label value 0 and 1
centroid_class_0 = np.mean(class_0_arr, axis=0)
centroid_class_1 = np.mean(class_1_arr, axis=0)
print("Centroid of class0 is:")
print(centroid_class_0)
print("Centroid of class1 is:")
print(centroid_class_1)
```

```python
#finding standard deviation of classes with label value 0 and 1
sd_class_0 = np.std(class_0_arr, axis=0)
sd_class_1 = np.std(class_1_arr, axis=0)
print("Standard deviation of class0 is:")
print(sd_class_0)
print("Standard deviation of class1 is:")
print(sd_class_1)
```

```python
from scipy.spatial.distance import euclidean
```

```python
#calculating interclass distance
euclidean_distance = euclidean(centroid_class_0, centroid_class_1)
print("Distance between centroids i.e interclass distance is:")
print(euclidean_distance)
```

```python
#calculating intradistance of class_0
intradistance_class_0 = np.mean(np.linalg.norm(class_0_arr - centroid_class_0, axis=1))
```

```python
print("Intradistance of class0 is:")
print(intradistance_class_0)
```

```python
#calculating intradistance of class_1
intradistance_class_1 = np.mean(np.linalg.norm(class_1_arr - centroid_class_1, axis=1))
```

```python
print("Intradistance of class1 is:")
print(intradistance_class_1)
```
